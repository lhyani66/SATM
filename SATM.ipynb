### Step 1: DATA CLEANING (TEXT PROCESSING)

```python
import re
import string

def clean_text(text):
    # Convert to lowercase
    text = str(text).lower()

    # Remove special characters, punctuation, and emojis
    text = re.sub(f'[{re.escape(string.punctuation)}]', '', text)
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra whitespaces
    text = re.sub(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F1E0-\U0001F1FF]', '', text)

    # Normalize common abbreviations
    text = text.replace('hw', 'homework')
    text = text.replace('ppt', 'presentation')
    text = text.replace('bio', 'biology')

    # Handle unclear or low-quality entries (simple replacement for 'idk', could be expanded)
    text = text.replace('idk', 'unknown')

    return text

df['Cleaned_Task_Text'] = df['Task Text'].apply(clean_text)

print("Original Task Text vs. Cleaned Task Text (first 5 rows):")
display(df[['Task Text', 'Cleaned_Task_Text']].head())
```

### Step 2: NLP PREPROCESSING

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize stop words and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def nlp_preprocess(text):
    # Tokenization
    tokens = word_tokenize(text)

    # Remove stop words and apply lemmatization
    processed_tokens = []
    for token in tokens:
        if token not in stop_words:
            processed_tokens.append(lemmatizer.lemmatize(token))

    return ' '.join(processed_tokens)

df['Processed_Task_Text'] = df['Cleaned_Task_Text'].apply(nlp_preprocess)

print("Original, Cleaned, and Processed Task Text (first 5 rows):")
display(df[['Task Text', 'Cleaned_Task_Text', 'Processed_Task_Text']].head())
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import pandas as pd

# Initialize separate encoders for Category and Importance
one_hot_encoder_category = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
one_hot_encoder_importance = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
label_encoder = LabelEncoder()

# One-hot encode 'Category' and 'Importance'
category_encoded = one_hot_encoder_category.fit_transform(df[['Category']])
importance_encoded = one_hot_encoder_importance.fit_transform(df[['Importance']])

# Create DataFrames from one-hot encoded features
category_df = pd.DataFrame(category_encoded, columns=one_hot_encoder_category.get_feature_names_out(['Category']))
importance_df = pd.DataFrame(importance_encoded, columns=one_hot_encoder_importance.get_feature_names_out(['Importance']))

# Ordinal encode 'Difficulty'
# Define the order for Difficulty
difficulty_order = ['Easy', 'Medium', 'Hard']
# Ensure the 'Difficulty' column is of 'category' dtype with the specified order
df['Difficulty'] = pd.Categorical(df['Difficulty'], categories=difficulty_order, ordered=True)
# Map categories to numerical values (0, 1, 2...)
df['Difficulty_Encoded'] = df['Difficulty'].cat.codes

print("Encoded Categorical Features (first 5 rows):")
display(pd.concat([category_df, importance_df, df[['Difficulty_Encoded']]], axis=1).head())
```

### Step 5: FEATURE ENGINEERING

```python
# Task length (number of words)
df['Task_Length'] = df['Processed_Task_Text'].apply(lambda x: len(x.split()))

# Urgency flag (if words like 'urgent', 'asap', 'tomorrow' appear)
urgent_keywords = ['urgent', 'asap', 'tomorrow', 'immediate']
df['Urgency_Flag'] = df['Cleaned_Task_Text'].apply(lambda x: 1 if any(keyword in x for keyword in urgent_keywords) else 0)

# Time pressure = Deadline / Time Taken
# Handle potential division by zero or very small Time Taken values to avoid NaNs or Infs
df['Time_Pressure'] = df['Deadline (days)'] / df['Time Taken (hours)'].replace(0, 0.001) # Replace 0 with a small number
df['Time_Pressure'] = df['Time_Pressure'].replace([float('inf'), -float('inf')], 1000) # Cap infinite values

print("Engineered Features (first 5 rows):")
display(df[['Task_Length', 'Urgency_Flag', 'Time_Pressure']].head())
```

### Step 7: HANDLE DATA QUALITY ISSUES

```python
import numpy as np

# Check for missing values
print("Missing values before handling:")
display(df.isnull().sum()[df.isnull().sum() > 0])

# For simplicity, if there are any missing values, we'll fill numerical NaNs with the mean
# and categorical NaNs with the mode. Here, we assume no NaNs were introduced by previous steps.
# If any appear, this is a robust way to handle them.
for col in df.columns:
    if df[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].fillna(df[col].mean(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)

print("\nMissing values after handling:")
display(df.isnull().sum()[df.isnull().sum() > 0])

# Detect outliers in 'Time Taken (hours)' using IQR method
Q1 = df['Time Taken (hours)'].quantile(0.25)
Q3 = df['Time Taken (hours)'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_count = df[(df['Time Taken (hours)'] < lower_bound) | (df['Time Taken (hours)'] > upper_bound)].shape[0]
print(f"\nNumber of outliers in 'Time Taken (hours)': {outliers_count}")

# Handle outliers by clipping (Winsorization)
df['Time Taken (hours)_clipped'] = df['Time Taken (hours)'].clip(lower=lower_bound, upper=upper_bound)

print("\n'Time Taken (hours)' original vs. clipped (first 5 rows with potential outliers):")
display(df[['Time Taken (hours)', 'Time Taken (hours)_clipped']].head())
```

### Step 10: MODEL TRAINING
Train three classifiers for Category and Importance, and three regressors for Time Taken. We re-fit TF-IDF only on the training split — correct ML practice.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# ── Features and targets ──────────────────────────────────────────
# At prediction time we only have the task text, so X = text only.
X_raw        = df['Processed_Task_Text']
y_category   = df['Category']            # 4 classes: Exam, Homework, Project, Quiz
y_importance = df['Importance']          # 3 classes: Low, Medium, High
y_time       = df['Time Taken (hours)']  # regression target: 1.0 – 7.0 hrs

# ── 80/20 split (stratified on Category for balanced classes) ─────
X_tr_raw, X_te_raw, y_cat_tr, y_cat_te = train_test_split(
    X_raw, y_category, test_size=0.2, random_state=42, stratify=y_category
)
_, _, y_imp_tr,  y_imp_te  = train_test_split(X_raw, y_importance, test_size=0.2, random_state=42, stratify=y_category)
_, _, y_time_tr, y_time_te = train_test_split(X_raw, y_time,       test_size=0.2, random_state=42, stratify=y_category)

# ── Fit TF-IDF on training data ONLY ─────────────────────────────
tfidf = TfidfVectorizer(max_features=500)
X_tr = tfidf.fit_transform(X_tr_raw)
X_te = tfidf.transform(X_te_raw)

print(f"Train : {X_tr.shape[0]} samples")
print(f"Test  : {X_te.shape[0]} samples")
print(f"Features: {X_tr.shape[1]}")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report

def train_classifiers(X_tr, X_te, y_tr, y_te, label):
    models = {
        'Naive Bayes'  : MultinomialNB(),
        'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
        'SVM'          : LinearSVC(random_state=42, max_iter=2000)
    }
    results = {}
    print(f"\n{'='*45}\n{label}\n{'='*45}")
    for name, model in models.items():
        model.fit(X_tr, y_tr)
        preds = model.predict(X_te)
        acc   = accuracy_score(y_te, preds)
        results[name] = {'model': model, 'accuracy': acc, 'preds': preds}
        print(f"  {name:20s} -> Accuracy: {acc:.4f}")
    best_name = max(results, key=lambda k: results[k]['accuracy'])
    best      = results[best_name]
    print(f"\n  Best: {best_name} ({best['accuracy']:.4f})")
    print(classification_report(y_te, best['preds']))
    return best['model'], best_name, best['accuracy']

best_cat_model, best_cat_name, cat_acc = train_classifiers(
    X_tr, X_te, y_cat_tr, y_cat_te, "CATEGORY CLASSIFICATION"
)
best_imp_model, best_imp_name, imp_acc = train_classifiers(
    X_tr, X_te, y_imp_tr, y_imp_te, "IMPORTANCE CLASSIFICATION"
)
```

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_absolute_error, r2_score

def train_regressors(X_tr, X_te, y_tr, y_te, label):
    models = {
        'Ridge Regression': Ridge(alpha=1.0),
        'Random Forest'   : RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
        'LinearSVR'       : LinearSVR(random_state=42, max_iter=2000)
    }
    results = {}
    print(f"\n{'='*45}\n{label}\n{'='*45}")
    for name, model in models.items():
        model.fit(X_tr, y_tr)
        preds = model.predict(X_te)
        mae   = mean_absolute_error(y_te, preds)
        r2    = r2_score(y_te, preds)
        results[name] = {'model': model, 'mae': mae, 'r2': r2}
        print(f"  {name:20s} -> MAE: {mae:.4f} hrs | R2: {r2:.4f}")
    best_name = min(results, key=lambda k: results[k]['mae'])
    best = results[best_name]
    print(f"\n  Best: {best_name} (MAE={best['mae']:.4f} hrs, R2={best['r2']:.4f})")
    return best['model'], best_name, best['mae']

best_time_model, best_time_name, time_mae = train_regressors(
    X_tr, X_te, y_time_tr, y_time_te, "TIME ESTIMATION (hours)"
)
```

### Step 12: IMPROVED MODEL TRAINING
Importance model now uses **TF-IDF + deadline** as features (deadline is the strongest real-world signal).
Time model now uses **TF-IDF + category** as features (category determines how long a task takes).

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from scipy.sparse import hstack, csr_matrix
import numpy as np

# ── Re-split with the new dataset ────────────────────────────────
X_raw        = df['Processed_Task_Text']
y_category   = df['Category']
y_importance = df['Importance']
y_time       = df['Time Taken (hours)']

X_tr_raw, X_te_raw, y_cat_tr, y_cat_te = train_test_split(
    X_raw, y_category, test_size=0.2, random_state=42, stratify=y_category
)
_, _, y_imp_tr,  y_imp_te  = train_test_split(X_raw, y_importance, test_size=0.2, random_state=42, stratify=y_category)
_, _, y_time_tr, y_time_te = train_test_split(X_raw, y_time,       test_size=0.2, random_state=42, stratify=y_category)

# ── TF-IDF fit on train only ──────────────────────────────────────
tfidf2 = TfidfVectorizer(max_features=500)
X_tr_tfidf = tfidf2.fit_transform(X_tr_raw)
X_te_tfidf = tfidf2.transform(X_te_raw)

# ── Deadline feature for importance model ─────────────────────────
deadline_scaler = StandardScaler()
dl_tr = df.loc[X_tr_raw.index, 'Deadline (days)'].values.reshape(-1, 1)
dl_te = df.loc[X_te_raw.index, 'Deadline (days)'].values.reshape(-1, 1)
dl_tr_scaled = deadline_scaler.fit_transform(dl_tr)
dl_te_scaled = deadline_scaler.transform(dl_te)

X_tr_imp = hstack([X_tr_tfidf, csr_matrix(dl_tr_scaled)])
X_te_imp = hstack([X_te_tfidf, csr_matrix(dl_te_scaled)])

# ── Category feature for time model ──────────────────────────────
le_cat = LabelEncoder()
cat_tr_enc = le_cat.fit_transform(df.loc[X_tr_raw.index, 'Category']).reshape(-1, 1)
cat_te_enc = le_cat.transform(df.loc[X_te_raw.index, 'Category']).reshape(-1, 1)

X_tr_time = hstack([X_tr_tfidf, csr_matrix(cat_tr_enc)])
X_te_time = hstack([X_te_tfidf, csr_matrix(cat_te_enc)])

print("Feature shapes:")
print(f"  Category model   : {X_tr_tfidf.shape}")
print(f"  Importance model : {X_tr_imp.shape}  (TF-IDF + deadline)")
print(f"  Time model       : {X_tr_time.shape} (TF-IDF + category)")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report

def train_classifiers_v2(X_tr, X_te, y_tr, y_te, label, allow_nb=True):
    models = {}
    if allow_nb:
        models['Naive Bayes']       = MultinomialNB()
    models['Random Forest']         = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
    models['SVM']                   = LinearSVC(random_state=42, max_iter=3000)
    models['Gradient Boosting']     = GradientBoostingClassifier(n_estimators=150, random_state=42)

    results = {}
    print(f"\n{'='*48}\n{label}\n{'='*48}")
    for name, model in models.items():
        Xtr = X_tr.toarray() if name == 'Gradient Boosting' else X_tr
        Xte = X_te.toarray() if name == 'Gradient Boosting' else X_te
        model.fit(Xtr, y_tr)
        preds = model.predict(Xte)
        acc   = accuracy_score(y_te, preds)
        results[name] = {'model': model, 'accuracy': acc, 'preds': preds}
        print(f"  {name:22s} -> Accuracy: {acc:.4f}")
    best_name = max(results, key=lambda k: results[k]['accuracy'])
    best      = results[best_name]
    print(f"\n  Best: {best_name} ({best['accuracy']:.4f})")
    print(classification_report(y_te, best['preds']))
    return best['model'], best_name, best['accuracy']

best_cat_model2, best_cat_name2, cat_acc2 = train_classifiers_v2(
    X_tr_tfidf, X_te_tfidf, y_cat_tr, y_cat_te,
    "CATEGORY  (TF-IDF only)", allow_nb=True
)
best_imp_model2, best_imp_name2, imp_acc2 = train_classifiers_v2(
    X_tr_imp, X_te_imp, y_imp_tr, y_imp_te,
    "IMPORTANCE (TF-IDF + deadline)", allow_nb=False
)
```

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_absolute_error, r2_score

def train_regressors_v2(X_tr, X_te, y_tr, y_te, label):
    models = {
        'Ridge Regression' : Ridge(alpha=1.0),
        'Random Forest'    : RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1),
        'LinearSVR'        : LinearSVR(random_state=42, max_iter=3000),
        'Gradient Boosting': GradientBoostingRegressor(n_estimators=150, random_state=42),
    }
    results = {}
    print(f"\n{'='*48}\n{label}\n{'='*48}")
    for name, model in models.items():
        Xtr = X_tr.toarray() if name == 'Gradient Boosting' else X_tr
        Xte = X_te.toarray() if name == 'Gradient Boosting' else X_te
        model.fit(Xtr, y_tr)
        preds = model.predict(Xte)
        mae   = mean_absolute_error(y_te, preds)
        r2    = r2_score(y_te, preds)
        results[name] = {'model': model, 'mae': mae, 'r2': r2}
        print(f"  {name:22s} -> MAE: {mae:.4f} hrs | R2: {r2:.4f}")
    best_name = min(results, key=lambda k: results[k]['mae'])
    best = results[best_name]
    print(f"\n  Best: {best_name} (MAE={best['mae']:.4f} hrs, R2={best['r2']:.4f})")
    return best['model'], best_name, best['mae']

best_time_model2, best_time_name2, time_mae2 = train_regressors_v2(
    X_tr_time, X_te_time, y_time_tr, y_time_te,
    "TIME ESTIMATION (TF-IDF + category)"
)
```
